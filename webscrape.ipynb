```python

pass_df = pass_df.iloc[1: , :]
```

```python
pass_df = pass_df.rename(columns={0:"Surname",
                   1:"FirstNames",
                   2:"Age",
                   3:"Boarded",
                   4:"Survived"})
```

```python
pass_df
```

```python
## simplify boarding
def encodeBoard(boarded):
    conditions = [boarded == 'Cherbourg',
                  boarded == 'Southampton',
                  boarded == 'Queenstown']

    values = ['C','S','Q']
    return np.select(conditions, values) 

pass_df['Board'] = pass_df.Boarded.apply(encodeBoard)
```

```python
# fixing titles
pass_df['FirstNames'] = pass_df['FirstNames'].str.replace(r'Mr', 'Mr. ')
pass_df['FirstNames'] = pass_df['FirstNames'].str.replace(r'Mrs', 'Mrs. ')
pass_df['FirstNames'] = pass_df['FirstNames'].str.replace(r'Miss', 'Miss. ')
pass_df['FirstNames'] = pass_df['FirstNames'].str.replace(r'Master', 'Master. ')
pass_df['FirstNames'] = pass_df['FirstNames'].str.replace(r'Mr. s', 'Mrs. ')
```

```python
# combining surname and firstnames
pass_df['Name'] = pass_df['Surname'] + ', ' + pass_df['FirstNames']
pass_df
```

```python
pass_df=pass_df.drop(['Boarded'],axis=1)
pass_df
```

```python
pass_df.to_csv('data/scraped_passengers.csv', encoding='utf-8')
```
